```python
#replace not available and unknown to nan
df_orders=pd.read_csv('orders.csv',na_values=['Not Available','unknown'])
```

```python
#rename Columns 
df_orders.rename(columns={'Ship Mode':'Ship_Mode','City':'city'})
df_orders.columns=(df_orders.columns.str.replace(' ','_')).str.lower()
```

```python
df_orders.head(10)
#calculate discount, sale_price and profit per product
df_orders['discount']=df_orders['list_price']*df_orders['discount_percent']/100
df_orders['sale_price']=df_orders['list_price']-df_orders['discount']
df_orders['profit']=abs(df_orders['sale_price']-df_orders['cost_price'])
```
